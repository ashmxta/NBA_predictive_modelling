```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
players = pd.read_csv('Players.csv')
player_stats = pd.read_csv('PlayerStatistics.csv')
team_histories = pd.read_csv('TeamHistories.csv')
team_stats = pd.read_csv('TeamStatistics.csv')
games = pd.read_csv('Games.csv')
league_schedule = pd.read_csv('LeagueSchedule24_25.csv')
```

```python
team_stats.columns
```

```python

# Preview the data (optional)
print("Players:")
print(players.head(), "\n")

print("Player Statistics:")
print(player_stats.head(), "\n")

print("Team Histories:")
print(team_histories.head(), "\n")

print("Team Statistics:")
print(team_stats.head(), "\n")

print("Games:")
print(games.head(), "\n")

print("League Schedule:")
print(league_schedule.head(), "\n")
```

```python


# ---- Visualization 2: Points vs Rebounds for Players ----
# (Assumes player_stats has columns 'Points' and 'Rebounds')
plt.figure(figsize=(8, 5))
plt.scatter(player_stats['points'], player_stats['reboundsTotal'], alpha=0.7)
plt.xlabel('Points')
plt.ylabel('Rebounds')
plt.title('Points vs. Rebounds')
plt.show()
```

```python

# ---- Visualization 3: Team Wins ----
plt.figure(figsize=(10, 6))
plt.bar(team_stats['teamName'], team_stats['win'], color='skyblue')
plt.xlabel('Team')
plt.ylabel('Wins')
plt.title('Team Wins')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
